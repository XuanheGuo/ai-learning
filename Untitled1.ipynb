```python
# ======================================
# Day 3:  + 
# ======================================

import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# ------------------------------
# Part 1: 
# ------------------------------

# y = 2x + 3
x_train = torch.tensor([[1.0],[2.0],[3.0],[4.0]])
y_train = torch.tensor([[5.0],[7.0],[9.0],[11.0]])

# 
model = nn.Sequential(
    nn.Linear(1, 1)
)

loss_fn = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)  # 
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=100, gamma=0.5)

loss_history = []
lr_history = []

for epoch in range(300):
    y_pred = model(x_train)
    loss = loss_fn(y_pred, y_train)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    scheduler.step()

    loss_history.append(loss.item())
    lr_history.append(optimizer.param_groups[0]['lr'])

# LossLR
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(loss_history, label="Loss")
plt.title("Loss ")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.subplot(1,2,2)
plt.plot(lr_history, label="Learning Rate", color="orange")
plt.title("")
plt.xlabel("Epoch")
plt.ylabel("LR")
plt.legend()
plt.show()

# ------------------------------
# Part 2: 
# ------------------------------

# sin(x)
x_train = torch.linspace(-2*torch.pi, 2*torch.pi, 200).unsqueeze(1)
y_train = torch.sin(x_train)

def train_model(model, epochs=2000, lr=0.01):
    loss_fn = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    for epoch in range(epochs):
        y_pred = model(x_train)
        loss = loss_fn(y_pred, y_train)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    return model

# 
activations = {
    "": nn.Identity(),
    "ReLU": nn.ReLU(),
    "Sigmoid": nn.Sigmoid(),
    "Tanh": nn.Tanh()
}

plt.figure(figsize=(10,5))
plt.scatter(x_train.numpy(), y_train.numpy(), label="", s=10)

colors = ['red','green','purple','orange']
for i, (name, act) in enumerate(activations.items()):
    if name == "":
        model = nn.Sequential(nn.Linear(1, 1))
    else:
        model = nn.Sequential(
            nn.Linear(1, 50),
            act,
            nn.Linear(50, 1)
        )
    trained_model = train_model(model)
    y_pred = trained_model(x_train).detach()
    plt.plot(x_train.numpy(), y_pred.numpy(), label=f"{name} ", color=colors[i])

plt.legend()
plt.title(" sin(x) ")
plt.show()
```
